### Linear Regression

```python
import tensorflow as tf
```

```python
# input Data
x_data = [[1],[2],[3]]
y_data = [[1],[2],[3]]
```

```python
# placeholder
X = tf.placeholder(tf.float32, shape=[None, 1]) 
    # shape = [None, 1]  데이터가 몇개 들어올지 알수 없고, 변동이 계속 발생하므로 none, 한번에 1개씩 들어오므로 1
Y = tf.placeholder(tf.float32, shape=[None, 1])
    # shape = [None, 1]  데이터가 몇개 들어올지 알수 없고, 변동이 계속 발생하므로 none, 한번에 1개씩 들어오므로 1
```

```python
# variable ( 변수 선언 및 임의의 값으로 초기화 )
W = tf.Variable(tf.random_normal(shape = [1])) # weight
b = tf.Variable(tf.random_normal(shape = [1])) # bias
```

```python
# model
model = X * W + b
```

```python
# cost
cost =  tf.reduce_sum(tf.square(model - Y))
```

```python
# train
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
# Session
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    # Training
    for step in range(2001):
        c, w_, b_, _ = sess.run([cost, W, b, train], feed_dict = {X:x_data, Y:y_data})
        print("step:%g, c:%g, w:%g, b:%g"%(step,c,w_,b_))
        
    # Testing
    print(sess.run(model, feed_dict= {X:[[4]]}))
    
```
